```python
import spacy

# Load SciSpacy model for medical text processing
nlp = spacy.load('en_core_sci_md')  # SciSpacy for medical terms

def preprocess(text):
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if not token.is_stop]
    return " ".join(tokens)

# Apply preprocessing to the abstract or body text fields
metadata_df['processed_text'] = metadata_df['abstract'].apply(preprocess)

print(metadata_df['processed_text'].head())
```
